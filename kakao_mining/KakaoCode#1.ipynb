```python
import pandas as pd
#메모장의 맨 위를 Time,ID:Message로 바꿔야함

friendly = 0 # 친하면 1 안친하면 0 입력

# 데이터 수집
kakaos = range(1,2)
pieces =[]

for i in kakaos:   
    path = 'kakao_%d.txt' % i
    kakao= pd.read_csv(path , error_bad_lines=False) # 읽어오기
    pieces.append(kakao)
       
kakao_ver1 = pd.concat(pieces, ignore_index=True) 
#kakao_ver1
```

```python

# 데이터 전처리 
length = len(kakao_ver1) # Nan 제거 전 길이 저장
kakao_ver1 = kakao_ver1.dropna() # Nan 제거
kakao_ver2 = kakao_ver1['ID:Message'] .str.split(' : ', expand =True) # : 으로 구분된 이름과 메시지를 따로 떼어냄
kakao_ver2.columns = ['Name','Message'] # 2개로 나누어졌다.
del kakao_ver1["ID:Message"] #기존에 있던 ID:Message 제거
kakao_ver2 = pd.concat([kakao_ver1, kakao_ver2], axis=1) # Time 과 Name, Message 이어 붙임

kakao_ver2
```

```python
#전처리 모듈1: 내가 보낸 카톡만 추출

def Myname(DataFrame, Myname): # 데이터프레임과 내 이름을 입력
    
    for i in range(1,length): 
        series = DataFrame.loc[i:i,'Name']    
        if series.values != ' '+Myname: # 내 이름이 아닌 rows는
            DataFrame = DataFrame.drop(i) # 없애버림
    return DataFrame
    
```

```python
# 전처리 모듈 3 : 오후일때 12시간 더 한다.
def OjunOhu(OjunOhu):
    
    if OjunOhu == '오후':
        hourplus = 12
    else:
        hourplus = 0
        
    return hourplus
```

```python
# 전처리 모듈 4: 발신일자가 한자리 수인경우 date의 위치가 달라지므로 구분
def dateChanger(real_time):
    
    if real_time[10:11] == '일': # 발신일이 한자리 수인지 확인
        # **아이폰의 경우 '일'이 아니라 '.' 을 입력한다.
        
        date = real_time[9:10] #date 를 저장
        hour = real_time[-5:-3] # 발신 시간 저장
        hour = OjunOhu(real_time[12:14]) + int(hour) #발신 시간이 오후면 +12hour / 오전이면 그대로 사용        
        
    else:
        date = real_time[9:11]
        hour = real_time[-5:-3]
        hour = OjunOhu(real_time[13:15]) + int(hour)        
      
    return date, hour
```

```python
#전처리 모듈 5: realtime을 epochtime으로 변경
import time
import datetime
def Timetounixtime(real_time):
    
    date, hour = dateChanger(real_time) #가공한 날짜와 시간 return
    
    stime = "%s/%s/%s" %(date,real_time[6:7], real_time[0:4]) # 일,월,년 저장
    sec=int(real_time[-2:])*60 +int(hour)*60*60 # 분과 시간을 초로 조정
    unixday = time.mktime(datetime.datetime.strptime(stime, "%d/%m/%Y").timetuple()) #시간차이를 알기 위해 epochtime 화
    unixtime = unixday + sec #epochtime은 초이므로 초를 그대로 더한다.
    return unixtime

```

```python
number= kakao_Time.index.values.tolist()
timelist = []
for i in number:
        timestamp = int(Timetounixtime(kakao_Time['Time'][i]))# 저장하기 위해 int 타입으로 변경
        timelist.append(timestamp) # 시간정보 저장
timelist.sort()
```

```python
timelen = len(timelist)-1 # 총길이
time_length = 0 # 얘기한 길이
sumtimegap = 0
for i in range(0,timelen):
    
    if (timelist[i+1]-timelist[i]) < 86400: #하루 이상 지난 경우는 제외
        
        sumtimegap = sumtimegap + (timelist[i+1]-timelist[i]) # 응답시간 간의 차이를 저장
        time_length = time_length+1
        
responsetime = sumtimegap/(time_length)

print('평균 응답 시간(초) =', responsetime) # 총 길이로 나눠줌
```
